```python
def create_array():
    lines=[]
    with open("2020_d2.txt") as f:
        line =f.readline()
        while line:
            line=line.strip('\n').split(' ')
            count = [int(x) for x in line[0].split('-')]
            character =line[1][0]
            password = line[2]
            lines.append([count,character,password])
            line = f.readline()
    return lines    

def counter(lines):
    count = 0
    for line in lines:
        first=line[0][0]-1
        second=line[0][1]-1
        character=line[1]
        password = line[2]
        try:
            if password[first] == character and password[second] != character: 
                count += 1
            elif password[first] != character and password[second] == character:
                count += 1
        except:
            pass
    return count

lines=create_array()
counter(lines)
```
